#**Summary**
<br>
This is a text story based game on the famous Lord of the Rings by J.R.R. Tolkien.\
I have used the Movie and the Book to make changes to the scenarios with a little bit\
of my own imagination. In this game you basically move on from one phase in terms of\
location to the next in 3 scenes. Scene_1 is from Bagend the house of Frodo & Bilbo\
to the borders of the Shire. Scene_2 is from the borders of the Shire to the town called\
Bree where the Prancing Pony is located. And Scene_3 is from the Prancing Pony to the\
lands of the Elves, Rivendell.\
The objective of this game is to have Frodo safely deliver the One Ring to Rivendell.\
